# Introduction

The finite element method is arguably the principle extant numerical technique for solving partial differential equations, particularly over non-trivial domain geometries, and can be interpreted to contain a number of other popular approachs (finite volume, finite difference) as special cases. 

::: {.notes}
I will attempt to motivate the method from a Functional Analytic viewpoint, keeping two goals in mind. I want to:

1. Provide enough generality to at least indicate avenues through which the method can be generalized beyond the Galerkin approach, and the typical hyperbolic problems that are used as examples;
2. Avoid getting bogged down in technicalities, particularly the details of constructing the linear algebraic systems that result from the discretization, which can often be avoided in practical applications anyway through the usage of the current generation of finite element software. 
:::

# 1. The functional-analytic viewpoint on PDEs
A **linear space** or **vector space** $V$ is a collection of objects that are closed under linear combination, i.e. if $x, y$ are in $V$ and $\alpha, \beta$ are real or complex numbers, then $\alpha x + \beta y$ is also in $V.$ 
The classic examples are the real and  complex Euclidean spaces $\mathbb{R}^n$ and $\mathbb{C}^n.$ 

::: {.notes}
Spaces of functions with a particular degree of regularity (Continuity, continuously differentiable, etc) can also be regarded as linear spaces. However, function spaces may lack certain of the nice algebraic and topological features of the Euclidean spaces. For example, they may not generally admit a finite basis, and may not be complete under a particular metric.  
:::



::: {.notes}
Function spaces may be endowed with a variety of norms, and picking a good norm may be problem dependent, since different norms on infinite dimensional spaces are not, in general, equivalent. They can also be equipped with inner products to create a geometric stucture. 
:::


## Geometric Structure
A typical choice of inner product for a function space might be 

\begin{align}
\langle f, g \rangle &:= \int_\Omega f g dx,\\
(f, g) &:= \int_\Omega f g + \nabla f \cdot \nabla g dx,
\end{align}
where $f, g$ are defined on the same domain $\Omega$ and $\nabla$ denotes the gradient operator. 
An inner product induces a norm, e.g.

\begin{align}
    \Vert f \Vert &:= \sqrt{\langle f, f \rangle }
\end{align}

defines a norm on many common function spaces.^[$f, g$ may have to be regarded as equivalence classes of functions in order to satisfy the reflexivity condition but this subtlety  can be safetly ignored in this discussion]

## Topological Structure

A **Hilbert space** is an inner product space which is complete with respect to the norm induced by the inner product. A function space might fail to be complete with respect to one norm, even if it is complete when equipped with another. 

The **dual space** of a normed space $V,$ denoted $V^*,$ is the collection of scalar-valued, continuous linear functions defined on $V.$ 


## Example

In the case of $\mathbb{R}^n,$ any scalar-valued linear mapping $x^*$ acts on vectors $x$ via 

\begin{align}
    x^*(x) = a_1 x_1 + a_2 x_2 + \ldots + a_n x_n
\end{align}

and so the dual can be identified with row vectors $\begin{bmatrix} a_1 a_2 \ldots a_n \end{bmatrix}$


I will refer to elements of $V^*$ as **functionals.**


::: {.notes}
If a functional $x^*$ in $(\mathbb{R}^n)^n$ is specified, then 

$$ x^*(x) = b $$
corresponds to a linear equation. Perhaps then an equation like 

$$ x^*(u) = f$$

if $x^*$ is in the dual of some function space will correspond to a PDE. 
:::



## Dual Spaces
In an inner product space $(V, [\cdot , \cdot])$ one way to define a functional is by a duality pairing: for $v$ in $V,$ we have that $[ v, \cdot ] =: v^*$ is in $V^*.$ That is, 
\begin{align}
v^*(u) = [ v, u ]
\end{align}
for $u$ in $V.$ 

## Dual Space Representation
This observation might motivate one to ask: are *all* linear functionals representable in this way? In other words, if $R_V \colon V \to V^*$ is defined by 
\begin{align}
    R_V(v) \mapsto [v, \cdot]
\end{align}

invertible?

## Theorem (Riesz Representation theorem) 

Suppose that $(H, [\cdot , \cdot ])$ is a (real) Hilbert space. Then $R_H$ is a linear isomorphism. 

In particular, $R_H$ is invertible. 

## Variational Form
The point of this whole discussion is that in order to make progress on some PDE, say 

\begin{align}
    A[u] = 0, \\
\end{align}

if we can establish a corresponding problem in the dual of some Hilbert space H, say 

\begin{align}
    R_H (u) = v^*. \\
\end{align}

for some $v^*$ in $H^*,$ we may conclude that $u = R_H^{-1}(v^*).$


## Example

Consider the PDE 

\begin{align}
    -\Delta u &= f \qquad \text{in } \Omega \\
    u &= 0 \qquad \text{ on }  \delta \Omega 
\end{align}

For some $\Omega \subset \mathbb{R}^n.$^[This PDE is called the Poisson equation where the operator $\Delta = \frac{\partial }{\partial x_1^2} + \ldots + \frac{\partial }{\partial x_n^2}.$]

::: {.notes}
Somehow, we must convert this problem into a problem in dual space. A typical approach is apply say $R_{L^2}$ to both sides of the PDE, obtaining $R_{L^2}(-\Delta u) = R_{L^2}(f)$ or, in other words, the so-called variational problem:

find $u$ so that 
\begin{align}
    \langle -\Delta u , v \rangle = \langle f, v \rangle
\end{align}
for all $v$ in $L^2(\Omega).$ Obviously this variational equation is a necessary condition for the strong Poisson problem, so if the dual problem turns out to be well-posed, we obtain a likely solution candidate.

Our variational statement is equivalent to 

\begin{align}
    \int_\Omega f v dx &= -\int_\Omega v\Delta u dx, \\
        &= \int_\Omega \nabla v \cdot \nabla u dx, 
\end{align}

if $v$ is assumed to have additional regularity such that integration by-parts may be performed. With $L(v) :=  \int_\Omega f v dx$ and $a(u, v) := \int_\Omega \nabla v \cdot \nabla u dx$ we rewrite again as 

find $u$ such that 
\begin{align}
    a(u, v) &= L(v), \qquad \text{for all $v$ in $H^1(\Omega)$} 
\end{align}

Note that $(H^1(\Omega), a)$ is not a Hilbert space, but $(H_0^1(\Omega), a)$ is. Conveniently, $u$ may also be assumed to reside in $(H_0^1(\Omega)).$ 

So, this variational problem can be thought of as 

\begin{align}
    R_{H_0^1(\Omega)}(u) = L,
\end{align}

and we conclude, via the Riesz representation theorem, that $u$ exists.
::: 

::: {.notes}
In this example, $a$ turned out to be an inner product on an appropriate function space. However, this turns out not to be necessary. Indeed, the following theorem can be established without too much difficulty:
:::

## Theorem (The Lax-Milgram theorem) 
Suppose that $a$ is a bilinear form, $L$ is a linear form defined on a Hilbert space $V$. If $a$ is 

1. **Bounded**, i.e. $a(u, v) \le C_B \Vert u \Vert \Vert v \Vert$ for all $u, v$ in $V,$
2. **Coercive**, i.e. $a(v, v) \ge C_C \Vert v \Vert^2,$

then there is a unique solution $u$ to the variational problem 

\begin{align}
    a(u, v) &= L(v) \qquad \text{for all $v$ in $V$}
\end{align}

and furthermore $\Vert u \Vert$ can be bounded by the initial data $L.$ 

::: {.notes}
Although the Lax-Milgram theorem is non-constructive, it provides the ingredients to develop an important case of the finite element method, called the Galerkin method. 

There are numerous refinements of this theorem in the literature (e.g. the assumptions may be relaxed to allow $u, v$ to lie in different spaces)
::: 


# 2. The finite element method

To develop a discrete problem, we instead consider a finite subspace 

\begin{align}
V^h &= \text{span}\{\phi_1, \ldots , \phi_N\} \subset V
\end{align}

and the discrete problem: find $u^h$ in $V^h$
such that 

\begin{align}
    a(u^h, v^h) &= L(v^h) \qquad \text{for all $v^h$ in $V^h$}
\end{align}

## Discrete problem 
Since $a$ is assumed to be bilinear and $L$ linear, if the coordinates of $u^h$ are given by $u^h = u_1 \phi_1 + \ldots + u_N \phi_N,$ then the discrete variational problem corresponds to the linear system 

\begin{align}
    \sum_{j=1}^N u_j  a(\phi_j, \phi_i) &= L(\phi_j), \qquad i=1,\ldots, N.
\end{align}

## Consistency 


Presumably, $V^h$ will be chosen to approximate $V$ as $N \to \infty.$ In the case where $a$ is an inner product, if $a(u, v^h) = L(v^h),$ then clearly 

$$a(u - u^h, v^h) = L(v^h) - L(v^h) = 0.$$

This means that the mapping $u \mapsto u_h$ defines an orthogonal projection from $V$ on the subspace $V^h,$ and that the $a$-norm of the error $u - u^h$ is minimal within $V^h.$ 

::: {.notes}
If $a$ is not an inner product, but satisfies the hypothesis of the Lax-Milgram theorem, a similar but slightly more technical result can be obtained. 

This consistency property - called **Galerkin Orthogonality** - frees one to focus on the representational task of designing $\{V_h\}$ to approximate $V$ efficiently. 

The usual choice is to select Lagrange polynomials of a particular degree, although splines and certain types of discontinuous bases both have applications as well. Generally a discrete basis requires a spatial discretization as well, conventionally a triangulation (or tetrahedralization), although other shapes are possible. 

:::

::: {.notes}

There is a plethora of software packages supporting various subsets of finite element methods. A particularly useful one is the Python FEniCS library. The idea behind FEniCS is that if the user can provide a reasonable variational formulation, it ought to be possible to more-or-less automatically assemble and solve the associated linear system.^[https://fenicsproject.org]

Although the project is still in development, and has certain drawbacks, I have found it to be quite useful in prototyping finite element models. 
:::

## Geometry setup 

```python
#FEniCS example
#| echo: true
from dolfinx import mesh, default_scalar_type
from dolfinx.fem import FunctionSpace, Constant, locate_dofs_geometrical, dirichletbc, Function, form, assemble_scalar
from dolfinx.fem.petsc import LinearProblem
from mpi4py import MPI
import numpy as np
import ufl

#mesh size 
h = 20
#create the domain
#comm argument is related to multiprocessing which isn't relevant here 
domain = mesh.create_unit_square(comm = MPI.COMM_WORLD, 
                                 nx = h,
                                 ny = h)

#test/trial function space V
# 1 is degree of piecewise polynomial 
V = FunctionSpace(mesh = domain, element = ("Lagrange", 1))
#define boundary conditions; here constant but could be a function 
uD = Constant(domain = domain, c = default_scalar_type(0))

#get the indices of the nodes with x coordinate 0 or 1
boundary_identifier = lambda x: (np.isin(x[0], [0 , 1]) | np.isin(x[1], [0 , 1]))
boundary_nodes = locate_dofs_geometrical(V=V, marker= boundary_identifier)
#describe boundary condtions 
boundary_condition = dirichletbc(value = uD, dofs = boundary_nodes, V=V)
```

## Variational Problem Formulation

```python
#define trial and test function

u = ufl.TrialFunction(V)
v = ufl.TestFunction(V)

#define forcing term
f = Function(V)
f_expr = lambda x: 2 * np.sin(2 * np.pi * x[1]) * (1 - 4 * np.pi**2 * x[0]**2 ) * np.sin(2 * np.pi * x[0]) + 4 * np.pi * x[0] * np.cos(2 * np.pi * x[0])
f.interpolate(f_expr )

A = (ufl.dot(ufl.grad(v), ufl.grad(u))) * ufl.dx
L = v * f * ufl.dx
```

## Assembly and linear solver

```python
problem = LinearProblem(A, L, bcs=[boundary_condition])
uh = problem.solve()
```

## Extensions

- Nonlinear problems can be solved by Newton-type iteration methods
- The assumptions of the Lax-Milgram Theorem can be weakend
- Many alternative types of spatial discretizations (e.g. different shapes or nonconforming meshes)
- Alternative families of elements may be useful in certain circumstances 
- Other types of variational problems can be considered
